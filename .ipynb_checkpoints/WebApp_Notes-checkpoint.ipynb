# Creates a subset of the data for webapp
- Split a small set of data
- Create a json call for each
- Aggregate results

```python
import pandas as pd
import json
```

### Creates tiny sets of data in /mnt/data/Fraud_Detection/webapp_sets/

```python
dataDF = pd.read_csv("/mnt/data/Fraud_Detection/creditcard.csv")
for i in range (0,4):
    chunkDF=dataDF.iloc[i*10:(i+1)*10,:-1].rename(columns={"Time":"Hour"})
    filename="/mnt/data/Fraud_Detection/webapp_sets/record_"+str(i)+".csv"
    chunkDF.to_csv(filename, encoding='utf-8', index=False)
chunkDF
```

### Creates individual jsons from /mnt/data/Fraud_Detection/webapp_sets/record_0.csv

```python
recordDF=pd.read_csv("/mnt/data/Fraud_Detection/webapp_sets/record_0.csv")
for t in range (1,9):
    chunkDF=recordDF.iloc[t]+recordDF.iloc[t,0]
    result = chunkDF.to_json(orient="index")
result
```
